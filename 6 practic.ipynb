```python
import matplotlib.pyplot as plt
import pandas as pd

games = pd.read_csv('game_info.csv')

plt.plot(games["id"].head(10), games["added_status_yet"].head(10), color="black", marker="o", markersize=7, label="Линейный график")
plt.plot(games["id"].head(10), games["added_status_owned"].head(10), color="pink", marker="o", markersize=10, label="Линейный график")
plt.plot(games["id"].head(10), games["added_status_beaten"].head(10), color="orange", marker="o", markersize=10, label="Линейный график")
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.grid()
plt.legend()
plt.show()
```

ЛИНЕЙНЫЕ ГРАФИКИ(им плохо, НО ОНИ РАБОТАЮТ) Прошу засчитать, т. к. в моем датасете трудно придумать нормально сравнение для линейного графика(да и вообще для любого графика)

```python


plt.bar(games["id"].head(10), games['added_status_yet'].head(10), color="yellow", label="Пример 1")
plt.grid()
plt.legend()
plt.show()

plt.bar(games["id"].head(10), games['rating_top'].head(10), color="purple", label="Пример 2")
plt.grid()
plt.legend()
plt.show()

plt.bar(games["id"].head(10), games['added_status_dropped'].head(10), color="red", label="Пример 3")
plt.grid()
plt.legend()
plt.show()
```

СТОЛБЧАТЫЕ ДИАГРАММЫ 

```python
plt.pie(games["added_status_dropped"].head(10), labels=games['id'].head(10))
plt.title("Пример 1")
plt.show()

plt.pie(games["added_status_owned"].head(10), labels=games['id'].head(10))
plt.title("Пример 2")
plt.show()
```

КРУГОВЫЕ ДИАГРАММЫ

```python
x = games['id'].head(10)
yet = games['added_status_yet'].head(10)
owned = games['added_status_owned'].head(10)
beaten = games['added_status_beaten'].head(10)

plt.scatter(x, yet, color="black")
plt.scatter(x, owned, color="pink")
plt.scatter(x, beaten, color="orange")
plt.show()
```

ДИАГРАММЫ РАССЕЯНИЯ

```python
plt.figure(figsize=(10, 6))
plt.hist(games["added_status_yet"], bins=30, color=["blue"], edgecolor="orange", alpha=0.7)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(games["added_status_owned"], bins=30, color=["blue"], edgecolor="orange", alpha=0.7)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(games["added_status_beaten"], bins=30, color=["blue"], edgecolor="orange", alpha=0.7)
plt.legend()
plt.show()
```
